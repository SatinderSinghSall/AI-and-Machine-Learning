# Writing Data to CSV & JSON Files:

```python
# Writing Data to CSV & JSON Files:

data3 = data.copy()

data3 = data3.drop_duplicates()
data3 = data3.fillna(0)

data3 = data3.sort_values("income")
data3 = data3.reset_index(drop=True)

data3.to_csv("sorted_data.csv")
```

# Group by & Aggregation of Data:
- groupby()
- mean / sum / max / min / count / etc
- agg()

```python
# Group by & Aggregation of Data:

data

data.groupby("country")["income"].mean()
data.groupby("country")["income"].min()
data.groupby("country")["income"].max()

data.groupby("gender")["income"].mean()
data.groupby("gender")["income"].max()

data.groupby("country")["income"].agg(["mean", "min", "max"])
```

# Melt & Pivot for Reshaping in Pandas:

```python
# Melt & Pivot for Reshaping in Pandas:

# --------------------------------------------
# 1. CREATE SAMPLE WIDE DATAFRAME
# --------------------------------------------
df = pd.DataFrame({
    "id": [1, 2],
    "Jan": [10, 20],
    "Feb": [12, 22],
    "Mar": [14, 24]
})

print("Original Wide DataFrame:\n", df, "\n")

# --------------------------------------------
# 2. MELT (Wide → Long Format)
# --------------------------------------------
df_melted = pd.melt(
    df,
    id_vars="id",          # Keep this column fixed
    var_name="Month",      # The melted column names move here
    value_name="Sales"     # The melted column values move here
)

print("After Melt (Long Format):\n", df_melted, "\n")

# --------------------------------------------
# 3. PIVOT (Long → Wide Format)
# --------------------------------------------
df_pivoted = df_melted.pivot(
    index="id",            # Rows
    columns="Month",       # New columns
    values="Sales"         # Values
)

print("After Pivot (Back to Wide):\n", df_pivoted, "\n")

# --------------------------------------------
# 4. OPTIONAL: FIX COLUMN INDEX AFTER PIVOT
# (Pivot creates a MultiIndex column sometimes)
# --------------------------------------------
df_pivoted = df_pivoted.reset_index()
df_pivoted.columns.name = None  # remove pivot column name

print("Cleaned Pivoted DataFrame:\n", df_pivoted)
```
