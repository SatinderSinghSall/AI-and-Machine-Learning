# Data Transformation - Feature Engineering:

```python
# Data Transformation - Feature Engineering:

data = pd.read_csv("raw_data.csv")
data

data2 = data.copy()
data2["tax"] = data2["income"].apply(lambda x : "20%" if x >= 50000 else "10%")

gender_map = {"Male": "M", "Female": "F", "Unknown": "U"}

data2["gender"] = data2["gender"].map(gender_map)

data2.assign(new_income = data2["income"] * 1.1)

data2
```
