# Machine Learning:
- __DBSCAN Clustering Algorithm.__
- Density-Based Spatial Clustering of Applications with Noise Algorithm.
- Unsupervised Machine Learning.

```python
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

## Load Dataset:

```python
iris = load_iris()

X = iris.data
y = iris.target
```

## Scalling Data:

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Visualize Data:

```python
sns.scatterplot(x = X_scaled[:, 0], y = X_scaled[:, 2])
```

## Applying DBSCAN Algorithm: Model

```python
dbscan = DBSCAN(
    eps = 0.8,
    min_samples = 5
)
```

```python
dbscan
```

## Generating Labels:

```python
labels = dbscan.fit_predict(X_scaled)
```

## Visualizing Labeled Data:

```python
sns.scatterplot(x = X_scaled[:, 0], y = X_scaled[:, 2], c = labels)
```

```python
labels
```

## Using DBSCAN Algorithm for non-linear datasets:

```python
X, y = make_moons(
    n_samples = 300,
    noise = 0.05,
    random_state = 42
)
```
