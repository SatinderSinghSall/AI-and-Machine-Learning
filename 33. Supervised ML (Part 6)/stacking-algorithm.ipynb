# Machine Learning: Algorithms.
- __Stacking__ Ensemble  Algorithm.
- Supervised Machine Learning.

```python
from sklearn.ensemble import StackingClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import StackingRegressor
from sklearn.metrics import r2_score
```

# Static Classifier:

## Classification Dataset:

```python
X, y = make_classification(
    n_samples = 500,
    n_features = 20,
    n_informative = 5,
    n_redundant = 2,
    random_state = 42
)
```

## Train, Test & Split:

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size = 0.3,
    random_state = 42
)
```

## Creating Individual Models for Training:
- Logistic Regression, Support Vector Classifier, and Tree Models.

```python
# Models:

lr = LogisticRegression()
svc = SVC()
dtc = DecisionTreeClassifier(max_depth = 3)
```

## Voting Classifier Model:

```python
meta_model = LogisticRegression()

stacking_classifier = StackingClassifier(
    estimators = [
        ("lr", lr),
        ("svc", svc),
        ("dtc", dtc)
    ],
    final_estimator = meta_model,
    cv = 5
)
```

```python
stacking_classifier.fit(X_train, y_train)
```

## Predictions:

```python
y_pred = stacking_classifier.predict(X_test)
```

## Evaluation Matrics:

```python
print(f"Accuracy Score: {accuracy_score(y_pred, y_test)}")
print(f"Classification Report: {classification_report(y_pred, y_test)} \n")
```

# Static Regressior

## Regressior Dataset:

```python
X, y = make_regression(
    n_samples = 500,
    n_features = 20,
    n_informative = 5,
    random_state = 42
)
```

## Train, Test & Split:

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size = 0.3,
    random_state = 42
)
```

## Creating Individual Models for Training:
- Logistic Regression, Support Vector Classifier, and Tree Models.

```python
# Models:

lin_reg = LinearRegression()
svr = SVR()
dtr = DecisionTreeRegressor(max_depth = 3)
```

## Voting Classifier Model:

```python
stacking_regressor = StackingRegressor(
    estimators = [
        ("lr", lin_reg),
        ("svr", svr),
        ("dtr", dtr)
    ]
)

stacking_regressor.fit(X_train, y_train)
```

## Predictions:

```python
y_pred = stacking_regressor.predict(X_test)
```

## Classification Dataset:

```python
print(f"r^2 Score: {r2_score(y_test, y_pred)} \n")
```
