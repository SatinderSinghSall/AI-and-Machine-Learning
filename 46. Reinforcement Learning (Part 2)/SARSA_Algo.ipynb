## Implementing SARSA (State-Action-Reward-State-Action) Algorithm.

```python
gamma = 0.99
alpha = 0.5
epsilon = 0.1
episodes = 5000
```

## Q-Table: Store Q-Values -> Action Values:

```python
Q = np.zeros((48, 4))
```

```python
Q
```

## Policy - Epsilon-Greedy: state => action

```python
def epsilon_greedy(state):
    if random.random() < epsilon:
        return env.action_space.sample() # random action => expore part
    else:
        return np.argmax(Q[state]) # exploit part (optimum action / maximum benefit)
```

## Implementing the SARSA (State-Action-Reward-State-Action) Algorithm.

```python
env = gym.make("CliffWalking-v1")

for episode in range(episodes):
    done = False # when the episode is ended
    state, _ = env.reset()
    action = epsilon_greedy(state)

    total_reward = 0
    episode_len = 0

    while not done:
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated
        next_action = epsilon_greedy(next_state)

        # SARSA Update:
        Q[state, action] = Q[state, action] + alpha * (
            reward + gamma * Q[next_state, next_action] - Q[state, action]
        )
        state = next_state
        action = next_action

        total_reward = total_reward + reward
        episode_len = episode_len + 1

    print(f"Episode Number: {episode} / 500 - Total Reward: {total_reward} & Episode Length: {episode_len}")
env.close()
```

## What did our Agent Learn?

```python
env = gym.make("CliffWalking-v1", render_mode = "human")

state, _ = env.reset()
done = False
total_reward = 0
episode_len = 0

while not done:
    action = np.argmax(Q[state])
    next_state, reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated

    state = next_state

    total_reward = total_reward + reward
    episode_len = episode_len + 1

print(f"Episode Number: {episode} / 500 - Total Reward: {total_reward} & Episode Length: {episode_len}")
env.close()
```
