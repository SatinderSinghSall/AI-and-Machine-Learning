# Series in Pandas:

```python
# Series in Pandas:

s = pd.Series([1, 2, 3, 4, 5])

print(s)
print(type(s))

# index
print(s[0])
print(s[3])
print(s.index)

print("\n")

# label:
s2 = pd.Series([22, 23, 24, 25, 26], index = ["Adam", "Eve", "Bob", "Charlie", "Nick"])
print(s2)
print(s2["Nick"])
print(s2["Charlie"])
print(s2.index)
```

# Series Properties:
- Homogeneous in Nature
- Vectorised Operations
- Handle Missing Values with NaN
- Mutable Values but Immutable Sizes

```python
# Series Properties:

s = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([10, 20, 30, 40, 50])

print(s + s2)
```

# Data Frame in Pandas:

```python
# Data Frame in Pandas:

info = {
    "Name" : ["Satinder", "Vaibhav", "Santosh"],
    "Age" : [22, 19, 22],
    "CGPA" : [6.9, 7.9, 8]
}

data_frame = pd.DataFrame(info)

print(data_frame, type(data_frame))
```

```python
data_frame
```

```python
print(data_frame.index)
print(data_frame.columns)
```

```python
data = [
    {"Name": "Satinder", "Age": 22, "CGPA": 6.9},
    {"Name": "Vaibhav", "Age": 19, "CGPA": 7.9},
    {"Name": "Santosh", "Age": 22, "CGPA": 8.0}
]

df = pd.DataFrame(data)
print(df)
```

```python
data = [
    ["Satinder", 22, 6.9],
    ["Vaibhav", 19, 7.9],
    ["Santosh", 22, 8.0]
]

df = pd.DataFrame(data, columns=["Name", "Age", "CGPA"])
print(df)
```

```python
data = (
    ["Satinder", "Vaibhav", "Santosh"],
    [22, 19, 22],
    [6.9, 7.9, 8]
)

df = pd.DataFrame({
    "Name": data[0],
    "Age": data[1],
    "CGPA": data[2]
})

print(df)
```
