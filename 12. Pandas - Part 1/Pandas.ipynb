# Series in Pandas:

```python
# Series in Pandas:

s = pd.Series([1, 2, 3, 4, 5])

print(s)
print(type(s))

# index
print(s[0])
print(s[3])
print(s.index)

print("\n")

# label:
s2 = pd.Series([22, 23, 24, 25, 26], index = ["Adam", "Eve", "Bob", "Charlie", "Nick"])
print(s2)
print(s2["Nick"])
print(s2["Charlie"])
print(s2.index)
```

# Series Properties:
- Homogeneous in Nature
- Vectorised Operations
- Handle Missing Values with NaN
- Mutable Values but Immutable Sizes

```python
# Series Properties:

s = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([10, 20, 30, 40, 50])

print(s + s2)
```

```python
data = [
    {"Name": "Satinder", "Age": 22, "CGPA": 6.9},
    {"Name": "Vaibhav", "Age": 19, "CGPA": 7.9},
    {"Name": "Santosh", "Age": 22, "CGPA": 8.0}
]

df = pd.DataFrame(data)
print(df)
```

# Filerating Data in Pandas

```python
# Filerating Data in Pandas:

data_frame[data_frame["aqi"] > 100]
data_frame[data_frame["aqi"] > 100]
data_frame[ (data_frame["aqi"] > 100) & (data_frame["temperature"] > 30)]
```

# Filerating Data using Queries in Pandas:

```python
# Filerating Data using Queries in Pandas:

aqi_value = 100
data_frame.query("aqi > 100 & temperature > 30") [["city", "aqi"]]
data_frame.query("aqi > @aqi_value & temperature > 30") [["city", "aqi"]]
```
