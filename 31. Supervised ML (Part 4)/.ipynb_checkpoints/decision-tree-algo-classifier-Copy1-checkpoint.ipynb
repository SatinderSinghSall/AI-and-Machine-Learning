# Machine Learning:
## Decision Tree Algorithm.
- __Decision Tree Algorithm.__ Using Classification.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
```

## Load Dataset:

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic
```

```python
titanic.describe
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
features = ["pclass", "sex", "fare", "embarked", "age"]
target = ["survived"]
```

```python
titanic.head()
```

## Handeling Missing Data:

```python
imp_median = SimpleImputer(strategy = "median")
titanic[["age"]] = imp_median.fit_transform(titanic[["age"]])

imp_freq = SimpleImputer(strategy = "most_frequent")
titanic[["embarked"]] = imp_freq.fit_transform(titanic[["embarked"]])
```

```python
titanic.head()
```

```python
titanic.isnull()
```

```python
titanic.isnull().sum()
```

```python
titanic.shape
```

# Data Encoding:

```python
titanic.head()
```

```python
label_encoder = LabelEncoder()
titanic["sex"] = label_encoder.fit_transform(titanic["sex"])
titanic["embarked"] = label_encoder.fit_transform(titanic["embarked"])
```

## Train, Test and Split:

```python
X = titanic[features]
y = titanic[target]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split (
    X, y,
    test_size = 0.3,
    random_state = 42
)
```

```python
X_train.head()
```

## Decision Tree Model:

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(f"Model Accuracy: {accuracy_score(y_test, y_pred)}")
```

## Tree Structure of Data:

```python
plt.figure(
    figsize=(18, 10)
)

plot_tree(
    model,
    feature_names = X.columns,
    class_names = ["Died" ,"Survived"],
    filled = True
)

plt.tight_layout()
plt.show()
```

```python
plt.figure(
    figsize=(18, 10)
)

plot_tree(
    model,
    feature_names = X.columns,
    class_names = ["Died" ,"Survived"],
    filled = True,
    max_depth = 2
)

plt.tight_layout()
plt.show()
```

## Decision Tree with Pre-Pruning:

```python
max_depth = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for depth in max_depth:
    model = DecisionTreeClassifier(max_depth = depth)
    model.fit(X_train, y_train)

    acc = model.score(X_test, y_test)
    print(f"For depth {depth}, accuracy: {acc}")

    if depth == 4:
        plt.figure(
        figsize=(18, 10)
    )
    
    plot_tree(
        model,
        feature_names = X.columns,
        class_names = ["Died" ,"Survived"],
        filled = True,
        max_depth = 2
    )
    
    plt.tight_layout()
    plt.show()
```

```python
max_depth = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for depth in max_depth:
    model = DecisionTreeClassifier(max_depth = depth)
    model.fit(X_train, y_train)

    acc = model.score(X_test, y_test)
    print(f"For depth {depth}, accuracy: {acc}")

    if depth == 4:
        plt.figure(
        figsize=(18, 10)
    )
    
    plot_tree(
        model,
        feature_names = X.columns,
        class_names = ["Died" ,"Survived"],
        filled = True,
    )
    
    plt.tight_layout()
    plt.show()
```

```python
max_depth = [2, 3, 4, 5, 6, 7, 8, 9, 10]
min_samples_split = [10, 15, 20, 25, 30]

for split in min_sample_split:
    model = DecisionTreeClassifier(max_depth = 4, min_samples_split = 10)
    model.fit(X_train, y_train)

    acc = model.score(X_test, y_test)
    print(f"For min sample split {split}, accuracy: {acc}")
```

```python
max_depth = [2, 3, 4, 5, 6, 7, 8, 9, 10]
min_samples_split = [10, 15, 20, 25, 30]

for split in min_sample_split:
    model = DecisionTreeClassifier(max_depth = 4, min_samples_split = 10)
    model.fit(X_train, y_train)

    acc = model.score(X_test, y_test)
    print(f"For min sample split {split}, accuracy: {acc}")

    if split == 10:
        plt.figure(
        figsize=(18, 10)
    )

    plot_tree(
        model,
        feature_names = X.columns,
        class_names = ["Died" ,"Survived"],
        filled = True,
    )

    plt.tight_layout()
    plt.show()
```
