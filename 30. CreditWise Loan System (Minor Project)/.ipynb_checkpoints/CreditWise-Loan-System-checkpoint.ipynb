```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

## Load Data:

```python
df = pd.read_csv("loan_approval_data.csv")
df.head()
df
df.info()
df.isnull().sum()
df.describe()
```

## Handle Missing Data / Values:

```python
categorical_cols = df.select_dtypes(include=["object"]).columns
numerical_cols = df.select_dtypes(include=["float64"]).columns # number
```

```python
categorical_cols
```

```python
numerical_cols
```

```python
categorical_cols.size + numerical_cols.size
```

```python
# Numerical NaN Filled with 'mean' values:

num_imp = SimpleImputer(strategy="mean")
num_imp.fit_transform(df[numerical_cols])
```

```python
df[numerical_cols] = num_imp.fit_transform(df[numerical_cols])
```

```python
df.head()
```

```python
# Catagorical NaN Filled with 'most frequent' values:

cat_imp = SimpleImputer(strategy="most_frequent")
df[categorical_cols] = cat_imp.fit_transform(df[categorical_cols])
```

```python
df
df.head()
df.isnull().sum()
```

## Expolatory Data Analysis: EDA
- How balanced our classes are.

```python
classes_count = df["Loan_Approved"].value_counts()

plt.pie(classes_count, labels=["No", "Yes"], autopct="%1.1f%%")
plt.title("Is Loan Approved or Not?")
```

```python
# Analyse Categories:

gender_count = df["Gender"].value_counts()
sns.barplot(gender_count)
```
