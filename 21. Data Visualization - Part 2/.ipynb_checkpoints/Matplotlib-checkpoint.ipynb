```python
import numpy as np
import matplotlib.pyplot as plt
```

# Histograms:

```python
np.random.seed(0)

scores = np.random.normal(70, 10, 100)
bins = [0, 30, 70, 90, 100]

plt.hist(
    scores,
    color="skyblue",
    edgecolor="black",
    bins=bins
)
plt.ylabel("# of Students")
plt.xlabel("Scores of Students")
```

```python
print(scores)
```

```python
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

plt.hist(
    marks,
    color="skyblue",
    edgecolor="black",
    bins=bins
)
plt.title("Distribution of Marks for Students.")
plt.ylabel("Number of Students")
plt.xlabel("Marks of Students")
```

# Multiple Datasets in Histograms:

```python
# Legit & Loud Transactions Distribution:

legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]

plt.hist(
    fraud_transactions,
    color="red",
    bins=10,
    label="Legit",
    edgecolor="black",
    alpha=0.5
)

plt.hist(
    legit_transactions,
    color="green",
    bins=10,
    label="Legit",
    edgecolor="black",
    alpha=0.5
)

plt.title("Legit & Loud Transactions Distribution.")
plt.xlabel("Amount (in $)")
plt.ylabel("Frequency in Transcations")
plt.grid()
plt.legend()
```

# Histograms: axvline

```python
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

plt.hist(
    marks,
    color="skyblue",
    edgecolor="black",
    bins=bins
)

plt.axvline(
    x=33,
    label="Passing Marks (33)",
    color="red",
    linewidth="2",
    linestyle="--"
)

plt.title("Distribution of Marks for Students.")
plt.ylabel("Number of Students")
plt.xlabel("Marks of Students")
plt.grid(True)
```
