# Seaborn Tutorial:
Seaborn is a Python data visualization library based on Matplotlib.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")

tips
```

```python
print(type(tips))
print(tips.head())
```

# Creating Plots with Seaborn:

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme(style="darkgrid")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme(style="whitegrid")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme(style="dark")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme(style="white")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme(style="ticks")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker"
)
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size"
)
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    style="smoker"
)
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line",
    hue="smoker",
    size="size",
    style="smoker"
)
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line",
)
```

# Relational Plots in Seaborn:
- Scatter Plots
- Line Plots

```python
tips = sns.load_dataset("tips")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
flights = sns.load_dataset("flights")

sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```
