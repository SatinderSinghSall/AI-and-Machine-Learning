```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

# Creating Plots with Seaborn:

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme(style="darkgrid")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme(style="whitegrid")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme(style="dark")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme(style="white")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme(style="ticks")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker"
)
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size"
)
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    style="smoker"
)
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line",
    hue="smoker",
    size="size",
    style="smoker"
)
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line",
)
```

# Relational Plots in Seaborn:
- Scatter Plots
- Line Plots

```python
tips = sns.load_dataset("tips")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
flights = sns.load_dataset("flights")

sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```

# Categorical Plots:

```python
print(tips.head())
```

```python
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

# Distribution Plots:

```python
penguins = sns.load_dataset("penguins")

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)
```

# Relation Plots in Seaborn - HeatMaps:
- Matrix Plots

```python
flights.head()
```

```python
flights
```

```python
flights = sns.load_dataset("flights")
flights.pivot(index="month", columns="year", values="passengers" )
```

```python
flights_pivot = flights.pivot(index="month", columns="year", values="passengers" )
sns.heatmap(
    flights_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d",
    linewidth=.5
)

plt.title("Passangers Heatmap.")
plt.tight_layout()
```
