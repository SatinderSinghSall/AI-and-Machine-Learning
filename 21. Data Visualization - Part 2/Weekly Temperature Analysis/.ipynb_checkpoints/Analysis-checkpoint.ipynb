# Weekly Temperature Analysis:
- City Temperature Visualisation using Python
- Multi-City Weekly Temperature Dashboard (Matplotlib)
- Weekly Weather Analysis with Python

Using Modern Matplotlib: Object-Oriented API
- Compare the temperature of cities over a week using subplots & loops.

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
cities = ["New York", "London", "Delhi", "Tokyo"]
temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

fig, axes = plt.subplots(2, 2)

city_num = 0
for i in range(2):        # rows
    for j in range(2):    # columns
        axes[i][j].plot(days, temperatures[city_num], marker='o')
        axes[i][j].set_title(cities[city_num])   #  FIX
        axes[i][j].grid(True)
        city_num += 1

fig.suptitle("Temperature in Cities over the Week")
fig.supylabel("Temperature")
fig.supxlabel("Days")
fig.tight_layout()

plt.show()
```
