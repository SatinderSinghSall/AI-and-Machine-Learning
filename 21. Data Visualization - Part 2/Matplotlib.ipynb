```python
import numpy as np
import matplotlib.pyplot as plt
```

# Histograms:

```python
np.random.seed(0)

scores = np.random.normal(70, 10, 100)
bins = [0, 30, 70, 90, 100]

plt.hist(
    scores,
    color="skyblue",
    edgecolor="black",
    bins=bins
)
plt.ylabel("# of Students")
plt.xlabel("Scores of Students")
```
