# Plot Methods: Using Real Data

```python
# Oscar Winner Movies data

oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(years, oscar_revenue)
plt.title("Oscer Movies Revinue in Each Year")
plt.xlabel("Years")
plt.ylabel("Revinue (in $M)")
```
