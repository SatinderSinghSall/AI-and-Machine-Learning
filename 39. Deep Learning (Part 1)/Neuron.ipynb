```python
inputs = torch.tensor([1.0, 2.0, 3.0]) # x1, x2 & x3 | 1D-Inputs
```

## Neuron Creation:

```python
neuron = nn.Linear(in_features = 3, out_features = 1)
```

## Output:

```python
output = neuron(inputs)
```
